# <span style="color:orange">**Semantic Segmentation in Roads**</span>

First in the following code the all libraries of the fastai should be imported:

```python
from fastai.vision.all import *
```

```python
path = untar_data(URLs.CAMVID_TINY)
dls = SegmentationDataLoaders.from_label_func(path, bs=8, fnames=get_image_files(path/"images"), label_func=lambda o: path/'labels'/f'{o.stem}_P{o.suffix}', codes=np.loadtxt(path/'codes.txt', dtype=str))
```

In the next line, I use the "unet_learner" to train the model:

```python
learn = unet_learner(dls, resnet34)
```

And in the next line, I decide how does the fastai fit my model:

```python
learn.fine_tune(8)
```

In  the next line, I show some sample of my trained data:

```python
learn.show_results(max_n=6, figsize=(7, 8))
```

In the next line, I got an image to do the prediction:

```python
import ipywidgets as widgets
uploader = widgets.FileUpload()
uploader
```

In the next line, I use the predict method to see the prediction result on my uploaded image:

```python
img = PILImage.create(uploader.data[0])
img, pred, probs = learn.predict(img)
img.show(figsize=(5,5))
```
